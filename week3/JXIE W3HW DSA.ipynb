```python
#---- xml parsing ------

tree = et.parse("week3_homework.xml")
root = tree.getroot()
```

```python
root[0]
```

```python
cols = []
for chd in root[0]:
    print(chd.tag,',',chd.text, ',',len(chd))
    if len(chd)>0:
        cols.append({chd.tag:chd.text})
```

```python
# now look at the event sequence
for chd in root[0][5]:
    print(chd.tag,',',chd.text, ',',len(chd))
```

```python
for chd in root[0][2][1]:
    print(chd.tag,',',chd.text, ',',len(chd))
```

```python
for chd in root[0][4][4]:
    print(chd.tag,',',chd.text, ',',len(chd))
```

```python
for i in range(4):
    d = {root[0][2][i][0].text:root[0][2][i][1].text}
    print(d)
    cols.append(d)

for i in range(5):
    d = {root[0][4][i][0].text:root[0][4][i][1].text}
    print(d)
    cols.append(d)
```

```python
# create event list into a list of dictionary
event_dict_list = []
for chd in root[0][5]:
    event_dict = {}
    for chdd in chd:
        event_dict.update({chdd.tag:chdd.text})
    event_dict_list.append(event_dict)
```

```python
xdf = pd.DataFrame(event_dict_list)
```
